```python
import numpy as np

x = np.linspace(-2**(1/2) + 2, 2**(1/2) + 2, 1000)
y = np.sqrt(2 - (x-2)**2) 
```

```python
lists1 = []
for a,b in zip(x,y):
    r = ((-2)*(b+2) - 2*a)/2
    if str(r) != 'nan':
        lists1.append(r)
        
lists2 = []
for a,b in zip(x,-y):
    r = ((-2)*(b+2) - 2*a)/2
    if str(r) != 'nan':
        lists2.append(r)
```

```python
print(max(lists1))

print(max(lists2))
```

```python
print(min(lists1))

print(min(lists2))
```
